#### 문자열 포매팅
- 문자열 안에 어떤 값을 삽입하는 방법

| 코드 | 설명 ㅣ 
| --- | --- |
| `%s` | 문자열(string) |
| `%d` | 정수(integer) |
| `%c` | 문자 1개(character) |
| `%f` | 부동 소수(floating-ponit) |
| `%o` | 8진수 |
| `%x` | 16진수 |
| `%%` | % 문자 자체 |

- format() 함수 사용
- f 문자열 포매팅(파이썬 3.6버전부터 지원)

```python
# %d => 숫자만 대입 가능
a = " I eat %d apples" % 10
```

```python
# %s => 어떤 값이든 대입 가능
# %s : 문자열 값

a = " I eat %s apples" % "five"
a
```

```python
a = "Error is %d%%" % 98
a
```

```python
a = " I ate %d apples, so I was sick for %s days" % (10, "three")
a
```

```python
# hi 문자열 출력 시 10개인 공간 사용
# 오른쪽 정렬
print("%10s" % "hi")

# 왼쪽 정렬
print("%-10sjane" % "hi")

# 0 : 길이에 상관하지 않음
# 소수점 네 번째 자리까지만
print("%0.4f" % 3.42134234)

# 전체 길이 10개인 문자열 공간에서 오른쪽으로 정렬
print("%10.4f" % 3.42134234)
```

```python
a = " I eat {} apples" .format(5)
a
```

```python
a = " I ate {} apples, so I was sick for {} days" .format(10, "three")
a = " I ate {0} apples, so I was sick for {1} days" .format(10, "three")
a
```

```python
name = "홍길동"
age = 30 

# 첫번째 방법
print("나의 이름은 %s 이고, 나이는 %d 입니다" % (name, age))
# 두번째 방법
print("나의 이름은 {} 이고, 나이는 {} 입니다" .format (name, age))
# f문자열 포매팅
print(f"나의 이름은 {name} 이고, 나이는 {age} 입니다")
print(f"나의 이름은 {name} 이고, 내년에 나이는 {age+1} 입니다")
```

#### 문자열 함수

```python
a = "hobby"

# 문자열안에 b라는 문자의 개수 세기
print(a.count('b'))

"test".count("t")
```

```python
# b위치 찾기
a = "Python is the best choice"

print(a.find('b'))

# 못찾으면 -1
a.find('k')
```

```python
# 대소문자 변경

print("hi".upper())
print("HI".lower())
```

```python
# 공백제거
# == : 같은가?

print('hi' == ' hi  ')

a = ' hi  '

# 왼쪽 공백 제거
print(a.lstrip())
# 오른쪽 공백 제거
print(a.rstrip())
# 양쪽 공백 제거
print(a.strip())
print('hi' == a.strip())
print(a)
```

```python
# 문자열 변경 : replace("바뀔문자", "변경할문자열")
a = "Life is too short"
print(a.replace("Life", "Your leg"))
print(a)
```
